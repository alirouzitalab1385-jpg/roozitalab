```python
import numpy as np
import matplotlib.pyplot as plt
def answer(list_1,storage,dpi_low=100,dpi_high=300):
     x=np.array(list_1)
     y=np.sin(x)
     plt.figure(figsize=(6,4))
     plt.plot(list_1,y,color='blue',linewidth=2,label='y=sin(x)')
     plt.title('chart sin(x) in range (-1,1)',fontsize=14,pad=15)
     plt.xlabel('x',fontsize=12)
     plt.ylabel('y',fontsize=12)
     plt.legend(loc='best',fontsize=10)
     plt.grid(True,alpha=0.3)
     plt.axhline(y=0,color='black',linewidth=0.5)
     plt.axvline(x=0, color='black',linewidth=0.5)
     plt.show()
     low=f"{storage}_low_{dpi_low}dpi.png"
     high=f"{storage}_high_{dpi_high}dpi.png"
     plt.savefig(low,dpi=dpi_low,bbox_inches='tight')
     plt.savefig(high,dpi=dpi_high,bbox_inches='tight')
     plt.close()
temp=np.linspace(-1,1,102)
list_1=temp[1:-1].tolist()
chart=answer(list_1,'sine_plot')
print(chart)
```
